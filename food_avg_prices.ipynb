```python
import pandas as pd
```

```python
df = pd.read_csv('food-price-index-march-2022-weighted-average-prices-csv-tables.csv')
```

```python
df.head()
df.tail()
```

```python
df.shape
```

<h3><font color=F7FF93>1. Checking dataframe information, finding Nan value, removing duplicate values

```python
df.info()
```

```python
df.isna().any()
df.isna().sum()
```

```python
df.loc[df['Data_value'].isna()]
```

```python
df.dropna(inplace=True)
df.shape
```

```python
df['Subject'].value_counts()
df['Group'].value_counts()
df['STATUS'].value_counts()
df['Series_title_1'].value_counts()
```

```python
df.drop(['Series_reference','STATUS','Subject','Group'], inplace=True, axis=1)
```

```python
df.shape
df.head()
```

<h3><font color=F7FF93>2. Converting 'Period' to datetime, 

```python
df.info()
```

```python
df['Period'] = df['Period'].astype(str)
df['Period'] = pd.to_datetime(df['Period'], format = "%Y"+"."+"%m").dt.strftime('%Y-%m')
df.head()
```

<h3><font color=F7FF93>3. Renaming columns

```python
df = df.rename(columns={'Data_value':'Price', 'Series_title_1':'Product'})
df.head(1)
```

<h3><font color=F7FF93>4. Finding top 20 popular products

```python
df['Product'].value_counts().head(20).index
```

```python
top20_prod = df[df['Product'].isin(df['Product'].value_counts().head(20).index)]
```

```python
top20_prod.head()
```

<h3><font color=F7FF93>5. Finding the most expensive product from the top 20 popular products

```python
import plotly.express as px
```

```python
top20_prod_mean = top20_prod.groupby('Product', as_index=False).mean()
top20_prod_mean.head()
```

<h3><font color=F7FF93>6. Average price of the 20 most expensive products

```python
ax = px.bar(  data_frame = top20_prod_mean , y = 'Product' , x = 'Price' , color = 'Price',
              title="Top 20 products sorted by price",
              facet_row_spacing=1,
              text='Price',
              orientation='h',
              color_continuous_scale='Viridis')

ax.update_yaxes(title='Products', title_standoff=10, categoryorder='total ascending',tickfont=dict({'size':10,'color':'gray'}))

ax.update_layout(showlegend=False,
                 margin=dict({'b':100,'pad':10,'r':200,'l':50}),
                 coloraxis_showscale=False)

ax.show()
```

<h3><font color=F7FF93>7. Finding out how susceptible were product prices to market fluctuations

```python
top20_prod_mean.Product.values
```

```python
df[df.Product.isin(top20_prod_mean.Product.values)]
```

```python
box = px.box( data_frame=df[df.Product.isin(top20_prod_mean.Product.values)],
              x='Product', 
              y='Price',
              color='Product',
              width=1600,
              height=1000,
              title='How susceptible were product prices to market fluctuations?')

box.update_layout(xaxis_title='Products',
                  yaxis_title='Product price',
                  grid_columns=20, 
                  showlegend=False )

box.update_xaxes( showgrid=True )

box.show()
```

<h3><font color=F7FF93>8. Price changes of the 5 most susceptible products over 20 years

```python
top20_prod_mean.Product.values
```

```python
top5 = ['Oranges, 1kg','Cheese - mild cheddar (supermarket only), 1kg','Butter - salted, 500g','Pizza, fresh or frozen, with any standard topping, each','Ham, sliced or shaved, 1kg', 'Mayonnaise, 380ml']
```

```python
top5_df = df[df['Product'].isin(top5)]
```

```python
top5_df.sort_values('Period', inplace=True)
```
